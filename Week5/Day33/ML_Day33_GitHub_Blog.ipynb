## 7. 博客黄金结构 — 照着写

> 好的技术博客 = 1 个核心观点 + 3 张图 + 10 行关键代码 + 20 分钟阅读时间

````python
# ============================================
# 技术博客"六段式"模板
# ============================================

print('博客六段式: 问题 -> 思想 -> 代码 -> 架构 -> 实验 -> 总结')
print()

template = '''
## 1. 问题: RNN 有什么问题?
- 顺序计算: 不能并行
- 梯度消失: 长程依赖难学

## 2. 核心思想: Self-Attention
- 每个词看所有词
- 公式: Attention(Q,K,V) = softmax(QK^T/sqrt(d))V

## 3. 手写实现 (关键代码)
def self_attention(Q, K, V):
    scores = Q @ K.T / math.sqrt(K.shape[-1])
    weights = softmax(scores, dim=-1)
    return weights @ V

## 4. 完整架构
- Scaled Dot-Product -> Multi-Head -> Transformer
- LayerNorm + 残差连接 + 位置编码

## 5. 实验结果
| 指标 | RNN | LSTM | Transformer |
| 训练时间 | 10min | 12min | 3min |
| 准确率 | 72% | 81% | 87% |

## 6. 总结
- Self-Attention 用并行解决了串行瓶颈
- Transformer = Self-Attention + 位置编码 + FFN
'''

print(template)

print('标题技巧:')
print('  数字: "3 种 Attention 机制对比"')
print('  问题: "Transformer 为什么需要 Multi-Head?"')
print('  对比: "GAN vs 扩散模型: 谁更强?"')
print('  实战: "Kaggle 房价预测: 从 0.5 到 0.2 的提分之路"')
````

## 9. GitHub Actions — 自动更新

> 用自动化让你的 GitHub 保持活跃

````python
# ============================================
# GitHub Actions YAML 配置
# ============================================

yaml_config = '''
# .github/workflows/update-readme.yml
name: Update README
on:
  schedule:
    - cron: '0 0 * * 0'
  workflow_dispatch:
jobs:
  build:
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v3
      - name: Generate README
        run: python scripts/generate_readme.py
      - name: Commit changes
        run: |
          git config --global user.name 'bot'
          git config --global user.email 'bot@github.com'
          git add README.md
          git commit -m 'auto: update README' || exit 0
          git push
'''

print('每周自动更新 README:')
print(yaml_config)

print()
print('保持提交节奏的方法:')
for t in [
    '每天写一行代码 -> commit -> 绿点不断',
    '把学习笔记也用 git 管理',
    '参与开源项目 -> 贡献 PR',
]:
    print(f'  {t}')
````
